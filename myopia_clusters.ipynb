```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Resources/myopia.csv')
df.head()
```

```python
df.drop('MYOPIC', axis=1, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(df)
X = pd.DataFrame(scaler.transform(df))
X.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.90)

X_pca = pd.DataFrame(pca.fit_transform(X))
print(X_pca.shape)
X_pca.head()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(learning_rate=50)
X_tsne = pd.DataFrame(tsne.fit_transform(X_pca))
print(X_tsne.shape)
X_tsne.head()
```

```python
plt.scatter(X_tsne[0], X_tsne[1])
plt.show()
```

```python
from sklearn.cluster import KMeans

inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(X_tsne)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(k)
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Based on the Elbow Curve above, it can be seen that the patients can be clustered at a value of 5, however the inertia value continues to decrease after this point although not as sharply. This is further supported by the scatter graph which identifies 5 clusters, although these are not very distinct and there are areas where patients overlap between clusters.
